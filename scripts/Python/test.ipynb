```python
count_of_sucess = []
for scenario in scenarios:
    
    response = requests.request("GET", scenario['url'], headers=headers, data=json.dumps(scenario['data']))
    sucess = 1 if(response.status_code) else 0
    count_of_sucess.append(sucess)

    with open(str('compare/'+str(scenario['nb'])+'.json'), 'w') as f:
        json.dump(response.json(), f)

    print(scenario['name'], sucess)

print(len(scenarios), sum(count_of_sucess), 'Sucess rate:', len(scenarios) / sum(count_of_sucess) * 100, '%')
test_rate.append(1)
```
